## 1.8.1 Total Sales by Category

This visualization compares total sales across the different product categories.

```python
sales_by_category = df.groupby("Category")["Sales"].sum()

sales_by_category.plot(kind="bar")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

### Observation

Technology generated the highest total sales, followed by Furniture and Office Supplies. This suggests that Technology products contribute the largest share of revenue.

## 1.8.2 Total Sales by Region

This visualization compares total sales generated across the different regions to identify the highest and lowest performing markets.

```python
sales_by_region = df.groupby("Region")["Sales"].sum()

sales_by_region.plot(kind="bar")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

### Observation

The chart compares total sales across the four regions. Regions with higher sales indicate stronger market performance and customer demand, while lower-performing regions may require additional marketing or sales strategies.

## 1.8.3 Total Profit by Category

This chart compares the total profit earned from each product category.

```python
profit_by_category = df.groupby("Category")["Profit"].sum()

profit_by_category.plot(kind="bar", color="green")
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.xticks(rotation=0)
plt.show()
```

### Observation

The visualization highlights which product categories contribute the most profit to the business. Comparing this with sales helps determine whether high sales also translate into high profitability.

```python
plt.figure(figsize=(8,6))

correlation = df[["Sales","Quantity","Discount","Profit"]].corr()

sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm",
            linewidths=0.5)

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

The heat map reveals the strength and direction of relationships among numerical variables. Positive values indicate that two variables increase together, while negative values suggest an inverse relationship. This analysis helps identify factors that influence sales and profitability.

# 2. Milestone 2: Data Preparation and Feature Engineering
This milestone focuses on cleaning, transforming, and engineering new features that improve data quality and provide additional insights for business decision-making.
New variables are created to uncover patterns that are not directly visible in the raw data set.

## 2.1 Data Grouping

### 2.1.1 Grouping Sales by Category

The data set is grouped according to product category to calculate the total sales generated by each category.

```python
sales_by_category = df.groupby("Category")["Sales"].sum()

sales_by_category
```

### 2.1.2 Grouping Sales by Region

The data set is grouped by region to calculate the total sales generated in each geographical region.

```python
sales_by_region = df.groupby("Region")["Sales"].sum()

sales_by_region
```

### 2.1.3 Grouping Profit by Category

The data set is grouped by product category to determine the total profit contributed by each category.

```python
profit_by_category = df.groupby("Category")["Profit"].sum()

profit_by_category
```

## 2.2 Data Transformation

### 2.2.1 Converting Order Date

The **Order Date** column is converted from a text format into a date time format.This is essential for performing time-series analysis.

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

## 2.3 Feature Engineering

### 2.3.1 Order Month

A new feature named **Order Month** is extracted from the Order Date column. This feature groups transactions by month, making it easier to identify seasonal trends and compare sales performance across different periods of the year.

```python
df["Order Month"] = df["Order Date"].dt.month_name()

df[["Order Date", "Order Month"]].head()
```

### 2.3.2 Profit Margin (%)

A new feature named Profit Margin (%) is calculated using the Sales and Profit columns. Profit margin measures profitability relative to sales, providing a more meaningful indicator of business performance and operational efficiency.

```python
df["Profit Margin (%)"] = (df["Profit"] / df["Sales"]) * 100

df[["Sales", "Profit", "Profit Margin (%)"]].head()
```

### 2.3.3 Sales Category

The continuous Sales variable is converted into three categorical groups: low, medium, and high. Categorizing sales values simplifies analysis by grouping similar transactions together, making it easier to identify purchasing patterns and compare different levels of sales performance.

```python
def sales_category(sales):
    if sales < 100:
        return "Low"
    elif sales < 500:
        return "Medium"
    else:
        return "High"

df["Sales Category"] = df["Sales"].apply(sales_category)

df[["Sales", "Sales Category"]].head()
```

## 2.4 Reproducible Workflow
The workflow followed in this milestone consists of:

1. Loading the cleaned data set.
2. Grouping data to summarize business performance.
3. Transforming date variables into date time format.
4. Creating new features through feature engineering.

# 3. Milestone 3: Visualization & Exploratory Analysis
This milestone focuses on visually exploring the prepared data set. 

## 3.1 Monthly Sales Trend

```python
monthly_sales = df.groupby("Order Month")["Sales"].sum()

month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

monthly_sales = monthly_sales.reindex(month_order)

plt.figure(figsize=(10,5))
monthly_sales.plot(kind="line", marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Order Month")
plt.ylabel("Total Sales")
plt.grid(True)

plt.show()
```

### Observation

The line graph shows how total sales change from one month to another. The fluctuations indicate that sales vary throughout the year, suggesting periods of higher and lower customer demand. These patterns can help the business identify seasonal trends and plan inventory and marketing strategies accordingly.

A line graph is the most suitable visualization because it is designed to display changes and trends over time. Since the data is organized by months, the line graph clearly illustrates increases and decreases in sales across the year, making trends easy to identify.

## 3.3 Average Profit Margin by Category

```python
profit_margin = df.groupby("Category")["Profit Margin (%)"].mean()

plt.figure(figsize=(8,5))
profit_margin.plot(kind="bar")

plt.title("Average Profit Margin by Category")
plt.xlabel("Category")
plt.ylabel("Average Profit Margin (%)")

plt.show()
```

### Observation

The bar chart compares the average profit margin across different product categories. It highlights which categories generate higher profitability relative to their sales and identifies those with lower profit margins.
A bar graph is appropriate because it provides a clear comparison between categories, making it easy to identify which category has the highest or lowest average profit margin.

## 3.4 Distribution of Profit Margin

```python
plt.figure(figsize=(8,5))

plt.hist(df["Profit Margin (%)"], bins=25)

plt.title("Distribution of Profit Margin")
plt.xlabel("Profit Margin (%)")
plt.ylabel("Frequency")

plt.show()
```

### Observation

The histogram shows how Profit Margin (%) values are distributed across all transactions. Most observations are concentrated within a particular range, while fewer transactions fall into extremely high or low profit margins, indicating the presence of variation in profitability.
A histogram is ideal for displaying the distribution of continuous numerical data since it helps identify the spread of values, concentration of observations and potential outliers that may not be visible in other chart types.

## 3.5 Sales versus Profit Margin

```python
plt.figure(figsize=(8,6))

plt.scatter(df["Sales"], df["Profit Margin (%)"])

plt.title("Sales vs Profit Margin")
plt.xlabel("Sales")
plt.ylabel("Profit Margin (%)")

plt.show()
```

### Observation

The scatter plot illustrates the relationship between Sales and Profit Margin (%). The points indicate that transactions with higher sales do not always generate higher profit margins, suggesting that profitability may also be influenced by factors such as discounts or operating costs.
A scatter plot is the most appropriate visualization for examining the relationship between two numerical variables. It helps identify trends, correlations, clusters, and unusual observations within the data.

## 3.6 Sales Category Distribution

```python
sales_category_distribution = df["Sales Category"].value_counts()

plt.figure(figsize=(6,5))
sales_category_distribution.plot(kind="bar")

plt.title("Sales Category Distribution")
plt.xlabel("Sales Category")
plt.ylabel("Number of Orders")

plt.show()
```

### Observation

The bar chart displays the number of transactions within each sales category (Low, Medium, and High). This provides an overview of customer purchasing behavior and shows which sales category contains the largest proportion of transactions.
A bar graph is suitable because Sales Category is a categorical variable. It allows easy comparison of the frequency of transactions across the different sales categories.

## 3.7 Correlation Heatmap

```python
plt.figure(figsize=(8,6))

correlation = df[
[
"Sales",
"Profit",
"Quantity",
"Discount",
"Profit Margin (%)"
]
].corr()

sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm",
            linewidths=0.5)

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

The heat map illustrates the strength and direction of relationships among the numerical variables. Positive correlations indicate that two variables tend to increase together, while negative correlations suggest that an increase in one variable is associated with a decrease in another. Weak correlations indicate little or no linear relationship.
A heat map is an effective visualization for displaying correlations among multiple numerical variables simultaneously. It enables patterns and relationships to be identified quickly without comparing variables individually.

## 3.8 Visualization Design Principles

The following visualization design principles were applied throughout this milestone:

- **Appropriate chart selection:** Each chart type was chosen based on the nature of the data and the analytical objective. For example, line graphs were used for trends over time, while bar graphs were used for comparing categories.

- **Clear titles and axis labels:** Every visualization includes a descriptive title and labelled axes to improve readability and ensure that the information is easy to interpret.

Applying these principles improves the clarity, consistency, and effectiveness of the visualizations.

## 3.9 Comparison of Visualization Strategies

- **Bar Graphs vs Line Graphs:** Bar graphs are most effective for comparing values across categories, while line graphs are better suited for displaying trends and changes over time.

- **Histogram vs Scatter Plot:** Histograms summarize the distribution of a single numerical variable, whereas scatter plots examine the relationship between two numerical variables, making them useful for identifying correlations and patterns.


## 3.10 Design Justification

The visualizations were selected based on the characteristics of the data and the insights required.

- The **line graph** was used to display monthly sales because it effectively illustrates trends over time.
- The **bar graphs** were chosen to compare categorical data such as product categories and sales categories.
- The **histogram** was selected to examine the distribution of Profit Margin (%) values.
- The **scatter plot** was used to investigate the relationship between Sales and Profit Margin (%).
- The **heat map** was chosen to summarize correlations among multiple numerical variables in a single visualization.
